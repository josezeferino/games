```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
data = pd.read_csv('./data/BrowserBased.csv')
data.head(20)
```

```python
data.boxplot (column= ['Star'], by='Type', figsize=(20,8))
plt.title('QUANTIDADE DE STAR POR GÊNERO')
plt.ylabel('STAR')
plt.show()
```

```python
#Resumo de informações em todas as colunas
data.describe()
```

```python
#Tipo de dado em cada coluna
data.dtypes
```

```python
data['Type'].value_counts()
```

```python
titulos_lancados = data['Type'].value_counts()
titulos_lancados.plot()

data['Type'].value_counts().plot()
```

```python
#Criando um gráfico utilizando apenas uma linha de código
data['Type'].value_counts().head(10).plot(kind='bar', figsize=(11,5), grid = False, rot=0, color='green')

#Enfeitando o gráfico. Abaixo, definimos um título
plt.title('Os 10 Gênero com mais títulos lançados')
plt.xlabel('Gênero') #Nomeando o eixo X, onde fica o nome dos videogames
plt.ylabel('Quantidade de jogos lançados') #Nomeando o eixo Y, onde fica a quantidade de jogos
plt.show() #Exibindo o gráfico
```

```python
#Os 10 jogos mais baixados
top_10_baixados = data[['Name', 'Star']].head(10).set_index('Name').sort_values('Star', ascending=True)
top_10_baixados.plot(kind='barh', figsize=(11,7), grid=False, color='darkred', legend=False)

plt.title('Os 10 jogos mais baixados')
plt.show()
```
